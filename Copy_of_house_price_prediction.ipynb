Impoer Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn as sk
from sklearn.datasets import fetch_california_housing    #get the data
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics  #for evaluation
```

importing boston house price dataset

```python
house_price_dataset = sk.datasets.fetch_california_housing()
```

```python
print(house_price_dataset)
```

```python
#load data to pandas dataframe
house_price_dataframe = pd.DataFrame(house_price_dataset.data, columns = house_price_dataset.feature_names)
```

```python
house_price_dataframe.head()
```

```python
#add the target column to the dataframe
house_price_dataframe['Price'] = house_price_dataset.target
```

```python
house_price_dataframe.head()
```

```python
#chek the number of rowa and columns
house_price_dataframe.shape
```

```python
#checking for missing values
house_price_dataframe.isnull().sum()
```

```python
#statistical measures of the dataset
house_price_dataframe.describe()
```

understanding the corrlation between the data

1 Positive Corrlation

2 negative Corrlation

```python
#constracting corrlation
corrlation = house_price_dataframe.corr()
```

```python
plt.figure(figsize = (10,10))
sns.heatmap(corrlation, cbar = True, square = True, fmt = '.1f', annot = True, annot_kws = {'size': 12}, cmap = 'coolwarm')
```

Splitting the data and target

```python
X = house_price_dataframe.drop(['Price'], axis = 1)
Y = house_price_dataframe['Price']
```

```python
print(X)
print(Y)
```

Splitting the data into training and test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state =2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

XGBoost Regressor it is descion tree algorithm

```python
#loading rhe model

model = XGBRegressor()
# training the model with X_train
model.fit(X_train, Y_train)
```

evaluating our model

Prediction on the training data

```python
#accuracy for prediction on training data
training_data_prediction = model.predict(X_train)
```

```python

print(training_data_prediction)
print(training_data_prediction.shape)
```

```python
#R square error
score_1 = metrics.r2_score(Y_train, training_data_prediction)

# mean absolute error
score_2 = metrics.mean_absolute_error(Y_train, training_data_prediction)

print("R square error: ", score_1)
print("Mean absolute error: ", score_2)
```

Visualizing the actual and predition price

```python
plt.scatter(Y_train, training_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual Price vs Predicted Price")
plt.show()
```

```python
#accuracy for prediction on test data
testing_data_prediction = model.predict(X_test)
print(testing_data_prediction)
```

```python
#R square error
score_1 = metrics.r2_score(Y_test, testing_data_prediction)

# mean absolute error
score_2 = metrics.mean_absolute_error(Y_test, testing_data_prediction)

print("R square error: ", score_1)
print("Mean absolute error: ", score_2)
```
