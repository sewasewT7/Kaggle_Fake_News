```python
#Im
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
```

```python
#Load the data
gold_price_data = pd.read_csv('/content/gld_price_data.csv')
gold_price_data.head()
```

```python
#get information about the dataset
gold_price_data.info()
gold_price_data.shape
```

```python
gold_price_data.describe()
```

```python
gold_price_data.isnull().sum()
```

```python
gold_price_data = gold_price_data.drop('Date',axis= 1)
correlation = gold_price_data.corr()
```

```python
sns.heatmap(correlation, annot=True)
plt.show()
```

```python
print(correlation['GLD'])
```

```python
sns.displot(gold_price_data['GLD'], color ='Blue')
plt.show()
```

```python
X = gold_price_data.drop('GLD', axis=1)
Y = gold_price_data['GLD']
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size =0.1, random_state=2)
```

```python
model = RandomForestRegressor(n_estimators=100)
model.fit(X_train, Y_train)
```

```python
gld_price_prediction = model.predict(X_train)
error_score = metrics.r2_score(Y_train, gld_price_prediction)
print("R squared error : ", error_score)
```

```python
model.fit(X_test, Y_test)
gld_price_prediction = model.predict(X_test)
error_score = metrics.r2_score(Y_test, gld_price_prediction)
print("R squared error : ", error_score)
```

```python
Y_test = list(Y_test)
```

```python
#visualization
plt.plot(Y_test, color='red', label = 'Actual Value')
plt.plot(gld_price_prediction, color='green', label='Predicted Value')
```
